```python
texto = input("Escribe una frase o un pequeño párrafo: ")
print("Texto original:", texto)
```

```python
texto = texto.lower()
print("Minúsculas:", texto)
```

```python
signos = ",.!?¿¡;:"
for   s in signos:
 texto = texto.replace(s, "")
print("Sin signos:", texto)
```

```python
texto = texto.strip() # quita espacios al inicio y final
```

```python
palabras = texto.split() # separa por espacios
```

```python
print("Palabras:", palabras)
```

```python
from collections import Counter
conteo = Counter(palabras)
print("Palabras más frecuentes:",conteo.most_common(3))
```
